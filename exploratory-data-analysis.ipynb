## Classification dataset

This is a classification dataset for determining whether the Loan Status of individuals, which is either Fully Paid or Charged Off. It has been taken from Kaggle datasets (https://www.kaggle.com/zaurbegiev/my-dataset) and although the documentation of where it was taken from is scarce, the readability of the features is good.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_classification = pd.read_csv('credit_train.csv', index_col=False)
```

```python
data_classification.head()
```

```python
data_classification.describe()
```

```python
data_classification.dropna(how="any", inplace=True)
# Dropping the nan values for now - only visualizing the data and there are enough observations 
# left even after dropping all nan values.
```

```python
print('T: ', 'Loan Status')
print('Rows: ', len(data_classification))
print('Columns: ', len(data_classification.columns))
```

```python
print('Distribution of non-numeric columns:\n')
print(data_classification['Loan Status'].value_counts(), '\n')
print(data_classification['Home Ownership'].value_counts(), '\n')
print(data_classification['Purpose'].value_counts(), '\n')
print(data_classification['Term'].value_counts(), '\n')
print(data_classification['Years in current job'].value_counts(), '\n')
```

```python
# Introducing new columns for numeric values of non-numeric data
data_classification['Fully Paid'] = (data_classification['Loan Status'] == 'Fully Paid')
data_classification['Charged Off'] = (data_classification['Loan Status'] == 'Charged Off')
data_classification['Short Term'] = (data_classification['Term'] == 'Short Term')
```

```python
for col in ['Fully Paid', 'Charged Off', 'Short Term']:
    data_classification = data_classification.astype({col:'int32'}, copy=False)
```

```python
data_classification.describe()
```

```python
for col in data_classification.columns:
    x = col.split()
    y = ''
    for i in x:
        y += i.lower()+'_'
    y = y[:-1]
    data_classification.rename(columns={col:y}, inplace=True)
```

```python
sns.countplot(data_classification['loan_status']);
```

We can observe how imbalanced the data is.

```python
data_classification.select_dtypes(include='number').hist(bins=15, figsize=(20, 20), layout=(5, 4));
```

Visualization shows interesting features. We'll look at them and their relationship with the loan status in more detail now.

```python
sns.scatterplot(x="annual_income", y="monthly_debt", hue="loan_status", data=data_classification);
```

```python
sns.scatterplot(x="credit_score", y="months_since_last_delinquent", hue="loan_status", data=data_classification);
```

### Preliminary observations for classification data:
1. The credit score for most of the participants is high. However this seems to have no effect on the ability of paying off the loan. However this might be a visualization issue, since we actually have more data for fully paid loans but this is not reflected in the plot.
2. The months since the person was last delinquent does not seem to have an effect on the loan payment.
3. A higher annual income even with a higher monthly debt shows a trend towards better fully paid loans.

## Regression dataset

This is a regression dataset for determining whether the price of houses. It has been taken from Kaggle datasets (https://www.kaggle.com/shivachandel/kc-house-data). Online property companies offer valuations of houses using machine learning techniques. The aim of this report is to predict the house sales in King County, Washington State, USA using Multiple Linear Regression (MLR). The dataset consisted of historic data of houses sold between May 2014 to May 2015.

```python
data_regression = pd.read_csv('kc_house_data.csv')
```

```python
data_regression.head()
```

```python
print('T: ', 'price')
print('Rows: ', len(data_regression))
print('Columns: ', len(data_regression.columns))
```

```python
sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 5]})
sns.distplot(
    data_regression['price'], norm_hist=False, kde=False, bins=20, color=['green'], hist_kws={"alpha": 1}
).set(xlabel='Sale Price', ylabel='Count');
plt.xticks(rotation=30)
plt.show()
```

Here also the data is pretty skewed. Let's look at the features.

```python
data_regression.select_dtypes(include='number').hist(bins=15, figsize=(20, 20), layout=(5, 4));
```

Owing to the nature of the data, the features are again quite interesting distribution wise. Let's see each in more detail.

```python
sns.jointplot(x=data_regression['bathrooms'], y=data_regression['price']);
```

```python
sns.jointplot(x=data_regression['bedrooms'], y=data_regression['price']);
```

```python
x = data_regression['bedrooms'] + data_regression['bathrooms']
sns.jointplot(x=x, y=data_regression['price']);
```

```python
sns.jointplot(x=data_regression['sqft_living'], y=data_regression['price']);
```

```python
sns.jointplot(x=data_regression['sqft_lot'], y=data_regression['price']);
```

### Preliminary observations for regression data:
1. The sqft_living feature seems to have a strong positive association with the house prices.
2. The combined number of bedrooms and bathrooms also show a positive association with the house prices.
3. The sale price is generally skewed towards 0 to 1 million, and therefore it might be harder to predict the prices of houses that might be more expensive than that.
